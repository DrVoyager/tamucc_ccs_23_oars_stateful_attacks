```python
import pandas as pd

# Define the file path


csv_file_path = 'NES_Untargeted_NOOP_CIFAR_Standard.xlsx'

# Load the CSV file into a DataFrame


df = pd.read_excel(csv_file_path)
#df = pd.read_csv(csv_file_path)

# Create separate lists for each column
sorted_list_first_element = df['SortedList_First_Element'].tolist()
sorted_list_second_element = df['SortedList_Second_Element'].tolist()

# Print the lists
print("First column values:", sorted_list_first_element)
print("Second column values:", sorted_list_second_element)
```

```python
len(sorted_list_first_element)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

# Assuming you have defined sorted_list_first_element and sorted_list_second_element
first_ele = sorted_list_first_element
second_ele = sorted_list_second_element

# Create a figure with a specified size
plt.figure(figsize=(14, 6))
plt.rcParams['figure.dpi'] = 300



# Plot the first element
plt.plot(
    range(1, len(first_ele) + 1),
    first_ele,
    marker="o",
    markersize=6,
    linestyle="-",
    color="b",
    linewidth=1,
    label="Largest Probability Score",
)

# Plot the second element
plt.plot(
    range(1, len(second_ele) + 1),
    second_ele,  # corrected from `seocnd_ele` to `second_ele`
    marker="x",
    markersize=6,
    linestyle="-",
    color="r",
    linewidth=1,
    label="Second Largest Probability Score",
)



plt.xlabel('Inference Requests', fontsize=22)
plt.ylabel('Probability Score', fontsize=22)



plt.title("Top-2 Probability Scores for NES Attack", fontsize=22)
plt.yticks(np.arange(0, 1.1, 0.2), fontsize=16)

# plt.xticks(np.arange(0, len(second_ele) + 1, 200), fontsize=16)


plt.xticks(np.concatenate([[1], np.arange(50, len(second_ele) + 1, 50)]), fontsize=16)




 # Gridlines
# for y in np.arange(0, 1.1, 0.1):
#         plt.axhline(y=y, color='lightgray', linestyle='-')
# for x in np.arange(0, len(second_ele) + 1, 50):
#         plt.axvline(x=x, color='lightgray', linestyle='-')



# Add a legend and grid
plt.grid(False)


save_plt_path = os.path.join(f"/content/NES_cifar10.png")
save_plt_path_pdf = os.path.join("/content/NES_cifar10.pdf")


# Show the plot

plt.legend(fontsize=20)
plt.savefig(save_plt_path)
plt.savefig(save_plt_path_pdf, format='pdf')

plt.show()
```

```python
import pandas as pd

# Define the file path
csv_file_path = 'Boundary_Cifar10.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)
```

```python
df.head()
```

```python
# Create separate lists for each column
sorted_list_first_element = df['Top 1 Probability'].tolist()
sorted_list_second_element = df['Top 2 Probability'].tolist()

# Print the lists
print("First column values:", sorted_list_first_element)
print("Second column values:", sorted_list_second_element)
```

```python
len(sorted_list_first_element)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

# Assuming you have defined sorted_list_first_element and sorted_list_second_element
first_ele = sorted_list_first_element
second_ele = sorted_list_second_element

# Create a figure with a specified size
plt.figure(figsize=(24, 6))
plt.rcParams['figure.dpi'] = 300



# Plot the first element
plt.plot(
    range(1, len(first_ele) + 1),
    first_ele,
    marker="o",
    markersize=6,
    linestyle="-",
    color="b",
    linewidth=1,
    label="Highest Output Activation Value",
)

# Plot the second element
plt.plot(
    range(1, len(second_ele) + 1),
    second_ele,  # corrected from `seocnd_ele` to `second_ele`
    marker="x",
    markersize=6,
    linestyle="-",
    color="r",
    linewidth=1,
    label="Second Highest Output Activation Value",
)



plt.xlabel('Inquiry Index', fontsize=22)
plt.ylabel('Predicted Score', fontsize=22)



plt.title("2-D Plot of Boundary Attack Probings")
plt.yticks(np.arange(0, 1.1, 0.1), fontsize=16)
# plt.xticks(np.arange(1, len(second_ele) + 1, 200), fontsize=16)
# Generate xticks at 1, 200, 400, 600, ...
plt.xticks(np.concatenate([[1], np.arange(10, len(second_ele) + 1, 10)]), fontsize=16)






 # Gridlines
for y in np.arange(0, 1.1, 0.1):
        plt.axhline(y=y, color='lightgray', linestyle='--')
for x in np.arange(0, len(second_ele) + 1, 20):
        plt.axvline(x=x, color='lightgray', linestyle='--')



# Add a legend and grid
plt.grid(True)

# Save the plot (self.image_number should be defined in your class)
# Commented out for now; replace with your own logic if necessary
save_plt_path = os.path.join(f"/content/graph1.png")
plt.savefig(save_plt_path)


# Show the plot




plt.legend(fontsize=18)

plt.show()
```
